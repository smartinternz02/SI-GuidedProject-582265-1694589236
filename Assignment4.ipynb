## Encoding the data

```python
dataset = pd.get_dummies(data=df,drop_first=True)
```

```python
dataset.info()
```

```python
# We see very high correlation between MonthlyIncome and JobLevel

correlation = dataset['MonthlyIncome'].corr(dataset['JobLevel'])

print(f"Correlation between MonthlyRate and JobLevel: {correlation}")
```

# Sepration of Dependent and Independent Variables

```python
X = dataset.drop(columns=['Attrition_Yes'],axis=1)
y = dataset['Attrition_Yes']
```

```python
X
```

```python
y
```

# Splitting in Testing and Training Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# MODEL BUILDING

# 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression(random_state=0)
```

```python
lr_clf.fit(X_train,y_train)
```

```python
# Making prediction
y_pred = lr_clf.predict(X_test)
```

### Load the metrics and Check the results

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
```

```python
acc_lr = accuracy_score(y_test,y_pred)
f1_lr = f1_score(y_test,y_pred)
prec_lr = precision_score(y_test,y_pred)
rec_lr = recall_score(y_test,y_pred)
```

```python
#Printing the result
results = pd.DataFrame([['LogisticRegression',acc_lr,f1_lr,prec_lr,rec_lr]],
                       columns=["Model","Accuracy","f1","Precision","Recall"])
results
```

```python
# Take a look at confusionn matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# Cross validation Score
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=lr_clf,X =X_train,y=y_train,cv=10)
accuracies.mean()*100
```

```python
accuracies.std()*100
```

# 2) Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
pred_dtc = dtc.predict(X_test)
pred_dtc
```

```python
y_test
```

```python
acc_dtc = accuracy_score(y_test,pred_dtc)
f1_dtc = f1_score(y_test,pred_dtc)
prec_dtc = precision_score(y_test,pred_dtc)
rec_dtc = recall_score(y_test,pred_dtc)
```

```python
result_dtc = pd.DataFrame([['Decision Tree',acc_dtc,f1_dtc,prec_dtc,rec_dtc]],
                       columns=["Model","Accuracy","f1","Precision","Recall"])
results = pd.concat([results,result_dtc],ignore_index=True)
results
```

```python
cm_dtc = confusion_matrix(y_test,pred_dtc)
cm_dtc
```

```python
# Cross validation Score
from sklearn.model_selection import cross_val_score
accuracies_dtc = cross_val_score(estimator=dtc,X =X_train,y=y_train,cv=10)
accuracies_dtc.mean()*100
```
